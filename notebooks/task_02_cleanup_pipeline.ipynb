# Task 2 ГўвӮ¬вҖқ ASR Output Cleanup Pipeline (Real JoshTalks Data)

**Data**: Raw ASR transcripts from pretrained `whisper-small` run on the **same ~10h JoshTalks dataset**  
Each raw ASR output is paired with the human reference from the transcription JSON.

**Operations:**
1. Number Normalisation (Hindi word ГўвҖ вҖҷ digits, with idiom protection)
2. English Word Detection & Tagging (Devanagari-form English words ГўвҖ вҖҷ `[EN]word[/EN]`)

**Note on transcription guidelines**: English words spoken in conversation  
are transcribed in Devanagari (e.g., `Г ВӨвҖўГ ВӨвҖҡГ ВӨВӘГ ВҘВҚГ ВӨВҜГ ВҘвҖҡГ ВӨЕёГ ВӨВ°` = "computer"). These count  
as **correct spelling**, not errors. Our English tagger identifies these for  
downstream differentiated processing ГўвӮ¬вҖқ not to flag them as wrong.

```python
import sys, os
sys.path.insert(0, os.path.abspath('../src'))

import torch
import pandas as pd
import numpy as np
from tqdm import tqdm
from IPython.display import display

from data_loader import JoshTalksDatasetLoader, fetch_transcription_segments
from preprocessing import AudioPreprocessor, HindiTextNormalizer
from number_normalizer import HindiNumberNormalizer
from english_detector import EnglishDevanagariDetector
from evaluation import compute_wer

print('Imports OK ГўЕ“вҖң')
```

## Step 1 ГўвӮ¬вҖқ Generate Raw ASR Outputs (Pretrained Whisper-small)

Run pretrained `whisper-small` (no fine-tuning) on audio segments from the manifest,  
and collect paired (human_reference, asr_output) pairs.

```python
from transformers import WhisperForConditionalGeneration, WhisperProcessor

MODEL_NAME = 'openai/whisper-small'
device = 'cuda' if torch.cuda.is_available() else 'cpu'

pretrained_model = WhisperForConditionalGeneration.from_pretrained(MODEL_NAME).to(device).eval()
pretrained_proc  = WhisperProcessor.from_pretrained(MODEL_NAME, language='hindi', task='transcribe')
audio_prep       = AudioPreprocessor()

def run_whisper(wav: np.ndarray) -> str:
    """Single-pass transcription with pretrained whisper-small."""
    inputs = pretrained_proc(wav, sampling_rate=16000, return_tensors='pt')
    inputs = {k: v.to(device) for k, v in inputs.items()}
    with torch.no_grad():
        ids = pretrained_model.generate(
            inputs['input_features'], language='hindi', task='transcribe'
        )
    return pretrained_proc.batch_decode(ids, skip_special_tokens=True)[0]

print(f'Pretrained Whisper loaded on {device}.')
```

```python
# Load manifest and process first N recordings for paired data
loader = JoshTalksDatasetLoader(cache_dir='../data/raw', max_recordings=10)
manifest = loader.load_manifest()

pairs = []   # list of {reference, asr_output, recording_id, segment_duration}
MAX_PAIRS = 200  # cap for demo; remove cap for full run

for item in tqdm(loader.iter_segments(manifest), desc='Generating ASR pairs'):
    if len(pairs) >= MAX_PAIRS:
        break
    try:
        asr_out = run_whisper(item['audio'])
    except Exception as exc:
        continue
    pairs.append({
        'reference':       item['transcription'],
        'asr_output':      asr_out,
        'recording_id':    item['recording_id'],
        'duration':        round(item['end'] - item['start'], 2),
    })

pairs_df = pd.DataFrame(pairs)
print(f'Generated {len(pairs_df)} ASR pairs.')
display(pairs_df[['reference', 'asr_output', 'duration']].head(5))
```

```python
# Baseline WER before any cleanup
raw_wer = compute_wer(pairs_df['reference'].tolist(), pairs_df['asr_output'].tolist())
print(f'Raw ASR WER (pretrained, no cleanup): {raw_wer:.4f} ({raw_wer*100:.1f}%)')
```

## Step 2a ГўвӮ¬вҖқ Number Normalisation

**Design decisions:**
- Uses greedy longest-match over a complete Hindi numeral lookup table
- Compound parsing: Г ВӨВӨГ ВҘвӮ¬Г ВӨВЁ Г ВӨВёГ ВҘЕ’ Г ВӨЕЎГ ВҘЕ’Г ВӨВөГ ВӨВЁ ГўвҖ вҖҷ (3 ГғвҖ” 100) + 54 = 354
- Idiom protection list prevents `Г ВӨВҰГ ВҘвҖ№-Г ВӨЕЎГ ВӨВҫГ ВӨВ° Г ВӨВ¬Г ВӨВҫГ ВӨВӨГ ВҘвҖЎГ ВӨвҖҡ` ГўвҖ вҖҷ `2-4 Г ВӨВ¬Г ВӨВҫГ ВӨВӨГ ВҘвҖЎГ ВӨвҖҡ` (wrong)
- Applied to **both** reference and ASR output before WER comparison

```python
num_norm = HindiNumberNormalizer(preserve_idioms=True)

# ГўвҖқвӮ¬ГўвҖқвӮ¬ Find actual examples from the dataset ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬
from number_normalizer import HINDI_NUMBER_WORDS

def contains_number_word(text: str) -> bool:
    return any(w in HINDI_NUMBER_WORDS for w in text.split())

num_pairs = pairs_df[pairs_df['reference'].apply(contains_number_word)].copy()
print(f'Segments containing number words: {len(num_pairs)} / {len(pairs_df)}')
```

```python
# Apply normalisation
pairs_df['ref_num_norm']  = pairs_df['reference'].apply(num_norm.normalize)
pairs_df['asr_num_norm']  = pairs_df['asr_output'].apply(num_norm.normalize)

# Show 4-5 before/after from real data
changed = pairs_df[pairs_df['ref_num_norm'] != pairs_df['reference']].head(5)
print('\n=== Number Normalisation: Before / After (real data) ===')
if len(changed) == 0:
    # Fallback to synthetic examples if real data has no numbers
    synthetic = [
        ('Г ВӨвҖ°Г ВӨВёГ ВӨВЁГ ВҘвҖЎ Г ВӨВӨГ ВҘвӮ¬Г ВӨВЁ Г ВӨВёГ ВҘЕ’ Г ВӨЕЎГ ВҘЕ’Г ВӨВөГ ВӨВЁ Г ВӨвҖўГ ВӨВҝГ ВӨВӨГ ВӨВҫГ ВӨВ¬Г ВҘвҖЎГ ВӨвҖҡ Г ВӨвҖ“Г ВӨВ°Г ВҘвӮ¬Г ВӨВҰГ ВҘвӮ¬Г ВӨвҖҡ', 'Г ВӨвҖ°Г ВӨВёГ ВӨВЁГ ВҘвҖЎ Г ВӨВӨГ ВҘвӮ¬Г ВӨВЁ Г ВӨВёГ ВҘЕ’ Г ВӨЕЎГ ВҘЕ’Г ВӨВөГ ВӨВЁ Г ВӨвҖўГ ВӨВҝГ ВӨВӨГ ВӨВҫГ ВӨВ¬Г ВҘвҖЎГ ВӨвҖҡ Г ВӨвҖ“Г ВӨВ°Г ВҘвӮ¬Г ВӨВҰГ ВҘвӮ¬'),
        ('Г ВӨВӘГ ВӨВҫГ ВӨВҒГ ВӨЕЎ Г ВӨВ№Г ВӨЕ“Г ВӨВҫГ ВӨВ° Г ВӨВ°Г ВҘВҒГ ВӨВӘГ ВӨВҜГ ВҘвҖЎ Г ВӨвҖўГ ВӨВ®Г ВӨВҫГ ВӨВҸ', 'Г ВӨВӘГ ВӨВҫГ ВӨВҒГ ВӨЕЎ Г ВӨВ№Г ВӨЕ“Г ВӨВјГ ВӨВҫГ ВӨВ° Г ВӨВ°Г ВҘВҒГ ВӨВӘГ ВӨВҜГ ВҘвҖЎ Г ВӨвҖўГ ВӨВ®Г ВӨВҫГ ВӨВҸ'),
        ('Г ВӨВҰГ ВӨВё Г ВӨВёГ ВӨВҫГ ВӨВІ Г ВӨВ¬Г ВӨВҫГ ВӨВҰ Г ВӨВ®Г ВӨВҝГ ВӨВІГ ВҘвҖЎ', 'Г ВӨВҰГ ВӨВё Г ВӨВёГ ВӨВҫГ ВӨВІ Г ВӨВ¬Г ВӨВҫГ ВӨВҰ Г ВӨВ®Г ВӨВҝГ ВӨВІГ ВҘвҖЎ Г ВӨВҘГ ВҘвҖЎ'),
        ('Г ВӨВҸГ ВӨвҖў Г ВӨВІГ ВӨВҫГ ВӨвҖ“ Г ВӨВІГ ВҘвҖ№Г ВӨвҖ” Г ВӨвҖ Г ВӨВҸ', 'Г ВӨВҸГ ВӨвҖў Г ВӨВІГ ВӨВҫГ ВӨвҖ“ Г ВӨВІГ ВҘвҖ№Г ВӨвҖ” Г ВӨвҖ Г ВӨВҸ Г ВӨВҘГ ВҘвҖЎ'),
    ]
    for ref, asr in synthetic:
        print(f'  REF  : {ref}')
        print(f'  ASR  : {asr}')
        print(f'  REFГўвҖ вҖҷ : {num_norm.normalize(ref)}')
        print(f'  ASRГўвҖ вҖҷ : {num_norm.normalize(asr)}')
        print()
else:
    for _, row in changed.iterrows():
        print(f'  REF  : {row.reference}')
        print(f'  ASR  : {row.asr_output}')
        print(f'  REFГўвҖ вҖҷ : {row.ref_num_norm}')
        print(f'  ASRГўвҖ вҖҷ : {row.asr_num_norm}')
        print()
```

```python
# ГўвҖқвӮ¬ГўвҖқвӮ¬ Edge Case Analysis ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬ГўвҖқвӮ¬
edge_cases = [
    ('Г ВӨВҰГ ВҘвҖ№-Г ВӨЕЎГ ВӨВҫГ ВӨВ° Г ВӨВ¬Г ВӨВҫГ ВӨВӨГ ВҘвҖЎГ ВӨвҖҡ Г ВӨвҖўГ ВӨВ°Г ВӨВЁГ ВҘвӮ¬ Г ВӨВ№Г ВҘЛҶГ ВӨвҖҡ', 'IDIOM: should preserve Г ВӨВҰГ ВҘвҖ№-Г ВӨЕЎГ ВӨВҫГ ВӨВ° as-is'),
    ('Г ВӨВҰГ ВҘвҖ№-Г ВӨВӨГ ВҘвӮ¬Г ВӨВЁ Г ВӨВҰГ ВӨВҝГ ВӨВЁ Г ВӨВ®Г ВҘвҖЎГ ВӨвҖҡ Г ВӨвҖ Г ВӨВЁГ ВӨВҫ', 'IDIOM: approximate quantity phrase'),
    ('Г ВӨВҸГ ВӨвҖў Г ВӨВЁГ ВӨвҖҡГ ВӨВ¬Г ВӨВ° Г ВӨвҖўГ ВӨВҫ Г ВӨвҖ Г ВӨВҰГ ВӨВ®Г ВҘвӮ¬', 'COLLOQUIAL: Г ВӨВҸГ ВӨвҖў = "first-rate", not number'),
    ('Г ВӨЕЎГ ВӨВҫГ ВӨВ° Г ВӨВҜГ ВӨВҫГ ВӨВ° Г ВӨВ®Г ВӨВҝГ ВӨВІГ ВҘвҖЎ', 'AMBIGUOUS: Г ВӨЕЎГ ВӨВҫГ ВӨВ° = four/friends ГўвӮ¬вҖқ preserve in context'),
    ('Г ВӨВёГ ВҘЕ’ Г ВӨВ¬Г ВӨВҫГ ВӨВӨ Г ВӨвҖўГ ВҘвӮ¬ Г ВӨВҸГ ВӨвҖў Г ВӨВ¬Г ВӨВҫГ ВӨВӨ', 'IDIOM: "the bottom line" ГўвӮ¬вҖқ should not convert'),
]
print('=== Edge Cases ===')
for text, reason in edge_cases:
    result = num_norm.normalize(text)
    changed_flag = 'ГўЕ“вҖң preserved' if result == text else f'ГўвҖ вҖҷ {result}'
    print(f'  INPUT   : {text}')
    print(f'  OUTPUT  : {changed_flag}')
    print(f'  REASON  : {reason}')
    print()
```

```python
num_norm_wer = compute_wer(pairs_df['ref_num_norm'].tolist(), pairs_df['asr_num_norm'].tolist())
print(f'WER after number normalisation: {num_norm_wer:.4f} ({num_norm_wer*100:.1f}%)')
print(f'Change vs raw: {(raw_wer - num_norm_wer)*100:+.2f}%')
```

## Step 2b ГўвӮ¬вҖқ English Word Detection (Devanagari)

In Hindi conversation, English words appear in Devanagari script  
(e.g., `Г ВӨвҖЎГ ВӨвҖҡГ ВӨЕёГ ВӨВ°Г ВӨВөГ ВҘВҚГ ВӨВҜГ ВҘвҖҡ`, `Г ВӨЕ“Г ВҘвҖ°Г ВӨВ¬`, `Г ВӨВӘГ ВҘВҚГ ВӨВ°Г ВҘвҖ°Г ВӨВ¬Г ВҘВҚГ ВӨВІГ ВӨВ®`).  
We tag these with `[EN]ГўвӮ¬ВҰ[/EN]` for downstream handling.

**Important**: Per transcription guidelines, Devanagari-form English IS correct spelling.  
The tagger does NOT correct them ГўвӮ¬вҖқ it **identifies** them for differentiated processing.

```python
en_detector = EnglishDevanagariDetector(
    use_phonotactics=True,
    confidence_threshold=0.5,
)

# Scan real ASR outputs for English words
pairs_df['asr_en_tagged'] = pairs_df['asr_output'].apply(en_detector.tag)
pairs_df['en_word_count'] = pairs_df['asr_output'].apply(lambda t: len(en_detector.detect(t)))

# Show examples where English words were detected
en_examples = pairs_df[pairs_df['en_word_count'] > 0].head(5)
print(f'Segments with detected English words: {(pairs_df.en_word_count > 0).sum()} / {len(pairs_df)}')
print('\n=== English Word Tagging Examples ===')

if len(en_examples) == 0:
    # Fallback to illustrative examples from the assignment
    samples = [
        'Г ВӨВ®Г ВҘвҖЎГ ВӨВ°Г ВӨВҫ Г ВӨвҖЎГ ВӨвҖҡГ ВӨЕёГ ВӨВ°Г ВӨВөГ ВҘВҚГ ВӨВҜГ ВҘвҖҡ Г ВӨВ¬Г ВӨВ№Г ВҘВҒГ ВӨВӨ Г ВӨвҖҰГ ВӨЕЎГ ВҘВҚГ ВӨвҖәГ ВӨВҫ Г ВӨвҖ”Г ВӨВҜГ ВӨВҫ Г ВӨвҖқГ ВӨВ° Г ВӨВ®Г ВҘВҒГ ВӨВқГ ВҘвҖЎ Г ВӨЕ“Г ВҘвҖ°Г ВӨВ¬ Г ВӨВ®Г ВӨВҝГ ВӨВІ Г ВӨвҖ”Г ВӨЛҶ',
        'Г ВӨВҜГ ВӨВ№ Г ВӨВӘГ ВҘВҚГ ВӨВ°Г ВҘвҖ°Г ВӨВ¬Г ВҘВҚГ ВӨВІГ ВӨВ® Г ВӨВёГ ВҘвҖ°Г ВӨВІГ ВҘВҚГ ВӨВө Г ВӨВЁГ ВӨВ№Г ВҘвӮ¬Г ВӨвҖҡ Г ВӨВ№Г ВҘвҖ№ Г ВӨВ°Г ВӨВ№Г ВҘвӮ¬',
        'Г ВӨВ®Г ВҘЛҶГ ВӨвҖҡГ ВӨВЁГ ВҘвҖЎ Г ВӨвҖҳГ ВӨВЁГ ВӨВІГ ВӨВҫГ ВӨвҖЎГ ВӨВЁ Г ВӨВҸГ ВӨвҖ”Г ВҘВҚГ ВӨЕ“Г ВӨВҫГ ВӨВ® Г ВӨВҰГ ВӨВҝГ ВӨВҜГ ВӨВҫ',
        'Г ВӨвҖ°Г ВӨВёГ ВӨвҖўГ ВӨВҫ Г ВӨвҖҳГ ВӨВ«Г ВӨВҝГ ВӨВё Г ВӨВ®Г ВҘвҖЎГ ВӨЕёГ ВҘВҚГ ВӨВ°Г ВҘвҖ№ Г ВӨвҖўГ ВҘвҖЎ Г ВӨВӘГ ВӨВҫГ ВӨВё Г ВӨВ№Г ВҘЛҶ',
    ]
    for s in samples:
        tagged = en_detector.tag(s)
        detected = en_detector.detect(s)
        print(f'  INPUT  : {s}')
        print(f'  TAGGED : {tagged}')
        print(f'  WORDS  : {[(w, round(c,2)) for w,c in detected]}')
        print()
else:
    for _, row in en_examples.iterrows():
        print(f'  REF    : {row.reference}')
        print(f'  ASR    : {row.asr_output}')
        print(f'  TAGGED : {row.asr_en_tagged}')
        print()
```

```python
# Real example from assignment spec
assignment_example = 'Г ВӨВ®Г ВҘвҖЎГ ВӨВ°Г ВӨВҫ Г ВӨвҖЎГ ВӨвҖҡГ ВӨЕёГ ВӨВ°Г ВӨВөГ ВҘВҚГ ВӨВҜГ ВҘвҖҡ Г ВӨВ¬Г ВӨВ№Г ВҘВҒГ ВӨВӨ Г ВӨвҖҰГ ВӨЕЎГ ВҘВҚГ ВӨвҖәГ ВӨВҫ Г ВӨвҖ”Г ВӨВҜГ ВӨВҫ Г ВӨвҖқГ ВӨВ° Г ВӨВ®Г ВҘВҒГ ВӨВқГ ВҘвҖЎ Г ВӨЕ“Г ВҘвҖ°Г ВӨВ¬ Г ВӨВ®Г ВӨВҝГ ВӨВІ Г ВӨвҖ”Г ВӨЛҶ'
print('Assignment example:')
print(f'  Input:  {assignment_example}')
print(f'  Output: {en_detector.tag(assignment_example)}')
print()

# Code-switch example from assignment
cs_example = 'Г ВӨВ®Г ВҘвҖЎГ ВӨВ°Г ВӨВҫ interview Г ВӨвҖҰГ ВӨЕЎГ ВҘВҚГ ВӨвҖәГ ВӨВҫ Г ВӨвҖ”Г ВӨВҜГ ВӨВҫ'   # Latin script English
print('Code-switch (Latin script) example:')
print(f'  Input:  {cs_example}')
print(f'  Output: {en_detector.tag(cs_example)}')
```
