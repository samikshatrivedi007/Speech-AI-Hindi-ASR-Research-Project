# Task 4 â€” Lattice-Based ASR Evaluation (Real 6-Model Data)

**Data**: [Task 4 Spreadsheet](https://docs.google.com/spreadsheets/d/1J_I0raoRNbe29HiAPD5FROTr0jC93YtSkjOrIglKEjU/edit#gid=1432279672)  
Columns: `segment_url_link, Human, Model H, Model i, Model k, Model l, Model m, Model n`

**Assignment spec**:
- Construct a lattice from all model outputs that captures valid alternatives
- Handle insertions, deletions, substitutions without unfairly penalising models
- When models agree against reference, trust model agreement over reference
- Compute WER per model using lattice-based matching
- Show WER decreases for unfairly penalised models; unchanged for others

**Alignment unit choice**: **Word-level**  
*Justification*: Hindi conversational ASR errors are predominantly word-level substitutions, not subword. Word-level alignment is interpretable and matches how human annotators reason about errors. Punctuation and case are normalised before alignment.

````python
import sys, os
sys.path.insert(0, os.path.abspath('../src'))

import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

from data_loader import Q4LatticeDataLoader
from lattice_builder import LatticeBuilder, LatticeWERComputer, build_and_evaluate
from preprocessing import HindiTextNormalizer
from jiwer import wer as standard_wer

print('Imports OK âœ“')
````

---
## 1. Load Real Data from Task 4 Spreadsheet

````python
q4_loader = Q4LatticeDataLoader()
df = q4_loader.load()

print(f'Rows loaded: {len(df)}')
print(f'Columns: {list(df.columns)}')
display(df.head(4))
````

````python
# Convert to evaluation records
records = q4_loader.to_evaluation_records(df)
MODEL_NAMES = list(records[0]['hypotheses'].keys())
print(f'Models in dataset: {MODEL_NAMES}')
print(f'Sample record:')
r = records[0]
print(f'  segment_url: {r["segment_url"]}')
print(f'  reference  : {r["reference"]}')
for m, h in r['hypotheses'].items():
    print(f'  [{m}]: {h}')
````

---
## 2. Text Normalisation

Before any WER computation, we normalise:
- Strip punctuation (`à¥¤`, `!`, `?`, `--`)
- Collapse whitespace
- Lowercase any Latin characters

This prevents penalising models for `à¥¤` vs `.` or `?` vs no punctuation.

````python
text_norm = HindiTextNormalizer()

def normalize(text: str) -> str:
    """Normalize text before WER computation."""
    # Remove Devanagari/Latin punctuation (common in model outputs)
    text = re.sub(r'[à¥¤à¥¥!?\.,;:\-â€“â€”"\(\)\[\]]+', ' ', text)
    text = text_norm.normalize(text)
    return text.strip()

# Apply to all records
for rec in records:
    rec['ref_norm'] = normalize(rec['reference'])
    rec['hyp_norm'] = {m: normalize(h) for m, h in rec['hypotheses'].items()}

# Show normalisation examples from real data
print('Normalisation examples from real data:')
for rec in records[:3]:
    print(f'  REF: {rec["reference"]:40s} â†’ {rec["ref_norm"]}')
    for m, h in list(rec['hyp_norm'].items())[:2]:
        orig = rec['hypotheses'][m]
        print(f'  [{m}]: {orig:40s} â†’ {h}')
    print()
````

---
## 3. Standard WER Per Model (Before Lattice)

Compute standard jiwer WER for each model vs. human reference.

````python
refs_norm = [rec['ref_norm'] for rec in records]

standard_wer_per_model = {}
for model in MODEL_NAMES:
    preds = [rec['hyp_norm'].get(model, '') for rec in records]
    wer_val = standard_wer(refs_norm, preds)
    standard_wer_per_model[model] = round(wer_val, 4)

std_df = pd.DataFrame([
    {'Model': m, 'Standard WER': v, 'Standard WER%': f'{v*100:.1f}%'}
    for m, v in standard_wer_per_model.items()
]).sort_values('Standard WER')

print('=== Standard WER (per model vs Human reference) ===')
display(std_df)
````

---
## 4. Build Word Lattice

**Algorithm:**
1. Select longest hypothesis as alignment backbone
2. Align each other hypothesis to backbone using Needleman-Wunsch (word-level)
3. At each aligned position, collect all alternatives into a `LatticeNode`
4. Insertions in any hypothesis create new nodes (with `''` in other hypotheses)

**Alignment unit**: Word  
**Gap penalty**: 1.0 (insertion/deletion cost = 1 word error)

````python
builder  = LatticeBuilder(insertion_penalty=1.0, deletion_penalty=1.0, substitution_penalty=1.0)
computer = LatticeWERComputer(agreement_discount=0.5, min_hypotheses_for_discount=3)

# Build lattice for each utterance
lattices = []
for rec in records:
    hyp_list = list(rec['hyp_norm'].values())
    lat = builder.build(hyp_list)
    lattices.append(lat)

# Show first lattice
print('=== Lattice for first utterance ===')
print(f'Reference: {records[0]["ref_norm"]}')
print(f'Positions:  {len(lattices[0])}')
for node in lattices[0].nodes:
    alts = sorted(a for a in node.alternatives if a)  # skip empty
    print(f'  Position {node.position}: {alts}')
````

````python
# Detailed view of all utterances
print('=== All Utterance Lattice Bins ===')
for i, (rec, lat) in enumerate(zip(records, lattices)):
    bins = [sorted(n.alternatives) for n in lat.nodes]
    print(f'\n[{i+1}] REF: {rec["reference"]}')
    print(f'    LATTICE: {bins}')
````

---
## 5. Lattice-Based WER Per Model

**When to trust model agreement over reference:**
- If â‰¥ 3 of 6 models produce the same word at a position, and it differs from reference â†’ apply 50% penalty discount
- Rationale: When a majority of independent models agree, the reference is likely a variant, not a ground truth error
- If only 1â€“2 models differ: no discount (could be coincidental errors)

````python
# Per-model lattice WER computation
hyp_lists_per_utterance = [
    list(rec['hyp_norm'].values()) for rec in records
]

lattice_wer_per_model = {}

for model in MODEL_NAMES:
    model_refs  = []
    model_preds = []
    total_edit  = 0.0
    total_ref   = 0

    for rec, lat, hyp_list in zip(records, lattices, hyp_lists_per_utterance):
        ref  = rec['ref_norm']
        pred = rec['hyp_norm'].get(model, '')

        result = computer.compute(ref, lat, hyp_list)
        total_edit += result['edit_distance']
        total_ref  += result['reference_length']

        model_refs.append(ref)
        model_preds.append(pred)

    lattice_wer_per_model[model] = round(total_edit / max(total_ref, 1), 4)

print('Lattice WER computed for all models.')
````

---
## 6. Full Comparison Table

````python
compare_rows = []
for model in MODEL_NAMES:
    std  = standard_wer_per_model[model]
    lat  = lattice_wer_per_model[model]
    diff = std - lat
    compare_rows.append({
        'Model':          model,
        'Standard WER':   f'{std:.4f}  ({std*100:.1f}%)',
        'Lattice WER':    f'{lat:.4f}  ({lat*100:.1f}%)',
        'Reduction':      f'{diff*100:+.2f}%',
        'Unfairly Penalised?': 'YES â†“' if diff > 0.02 else ('NO' if diff <= 0 else 'MINOR'),
    })

compare_df = pd.DataFrame(compare_rows).sort_values('Lattice WER')
print('=== WER Comparison: Standard vs Lattice-Based ===')
display(compare_df)
````

````python
# â”€â”€ Per-utterance drill-down â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print('=== Per-Utterance Analysis (Standard vs Lattice WER) ===')
for i, (rec, lat) in enumerate(zip(records, lattices)):
    hyp_list = hyp_lists_per_utterance[i]
    result = computer.compute(rec['ref_norm'], lat, hyp_list)
    print(f'\n[{i+1}] REF:      {rec["reference"]}')
    for m, h in rec['hypotheses'].items():
        std_w  = round(standard_wer([rec['ref_norm']], [rec['hyp_norm'][m]]), 3)
        print(f'     [{m}]: {h}')
        print(f'             STD WER={std_w:.3f}  LATTICE WER={result["lattice_wer"]:.3f}')
    print(f'     Lattice bins: {[sorted(n.alternatives) for n in lat.nodes]}')
````

````python
# â”€â”€ Visualisation â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
fig, ax = plt.subplots(figsize=(12, 5))

x = np.arange(len(MODEL_NAMES))
width = 0.35

std_vals = [standard_wer_per_model[m]*100 for m in MODEL_NAMES]
lat_vals = [lattice_wer_per_model[m]*100  for m in MODEL_NAMES]

bars1 = ax.bar(x - width/2, std_vals, width, label='Standard WER', color='#e74c3c', alpha=0.85)
bars2 = ax.bar(x + width/2, lat_vals, width, label='Lattice WER',  color='#2ecc71', alpha=0.85)

for b, v in zip(bars1, std_vals):
    ax.text(b.get_x()+b.get_width()/2, b.get_height()+0.3, f'{v:.1f}',
            ha='center', va='bottom', fontsize=8, fontweight='bold')
for b, v in zip(bars2, lat_vals):
    ax.text(b.get_x()+b.get_width()/2, b.get_height()+0.3, f'{v:.1f}',
            ha='center', va='bottom', fontsize=8, fontweight='bold')

ax.set_xticks(x)
ax.set_xticklabels(MODEL_NAMES)
ax.set_ylabel('WER (%)')
ax.set_title('Task 4 â€” Standard vs Lattice-Based WER per Model')
ax.legend()
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('../outputs/task_04_lattice_wer_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
````

## 7. Agreement Discount Ablation

````python
discount_results = []
for disc in [0.0, 0.25, 0.5, 0.75, 1.0]:
    comp = LatticeWERComputer(agreement_discount=disc, min_hypotheses_for_discount=3)
    avg_lat_wer = []
    for model in MODEL_NAMES:
        tot_edit, tot_ref = 0.0, 0
        for rec, lat, hyp_list in zip(records, lattices, hyp_lists_per_utterance):
            r = comp.compute(rec['ref_norm'], lat, hyp_list)
            tot_edit += r['edit_distance']
            tot_ref  += r['reference_length']
        avg_lat_wer.append(tot_edit / max(tot_ref, 1))
    discount_results.append({'discount': disc, 'avg_lattice_wer': np.mean(avg_lat_wer)})

disc_df = pd.DataFrame(discount_results)
plt.figure(figsize=(8, 4))
plt.plot(disc_df['discount'], disc_df['avg_lattice_wer']*100, 'o-', color='#3498db', lw=2, ms=8)
mean_std = np.mean(list(standard_wer_per_model.values()))*100
plt.axhline(mean_std, color='#e74c3c', ls='--', label=f'Mean Standard WER ({mean_std:.1f}%)')
plt.xlabel('Agreement Discount')
plt.ylabel('Avg Lattice WER (%)')
plt.title('Effect of Agreement Discount on Lattice WER')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('../outputs/task_04_discount_ablation.png', dpi=150, bbox_inches='tight')
plt.show()

display(disc_df)
````

## 8. Pseudocode Summary

```
ALGORITHM: Lattice-Based WER

INPUT:  reference R, hypotheses H = {H1, H2, ..., Hn}

STEP 1 â€” Build Lattice L:
  backbone â† longest hypothesis in H
  L â† [LatticeNode({backbone[i]}) for i in range(len(backbone))]
  FOR each Hi in H (skip backbone):
    (aligned_backbone, aligned_Hi) â† NeedlemanWunsch(backbone, Hi, word-level)
    FOR each (b, h) in zip(aligned_backbone, aligned_Hi):
      IF b != "": L[position(b)].add(h)
      ELSE:       L.insert_new_node({h, ""})

STEP 2 â€” Flexible WER(R, L):
  dp[0][0..len(L)] = insertions
  dp[0..len(R)][0] = deletions
  FOR i in 1..len(R):
    FOR j in 1..len(L):
      is_match â† R[i] in L[j].alternatives
      sub_cost â† 0 if is_match else compute_cost(R[i], L[j], H)
        WHERE compute_cost applies agreement_discount if |models agreeing| >= threshold
      dp[i][j] = min(dp[i-1][j-1] + sub_cost, dp[i-1][j] + 1, dp[i][j-1] + 1)
  lattice_WER â† dp[len(R)][len(L)] / len(R)

STEP 3 â€” Per-Model WER:
  FOR each model Hi:
    standard_WER(Hi) â† jiwer.wer(R, Hi)  # single hypothesis
    lattice_WER(Hi)  â† Lattice-WER(R, L) # shared lattice from all models
    IF lattice_WER < standard_WER: model was unfairly penalised by rigid reference
```
